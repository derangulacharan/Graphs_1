```python
y=np.array([10,135,25,19])
plt.pie(y)
plt.show()
```

```python
y=np.array([10,35,25,19])
mylabels=["Cherries","Apples","Bananas","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
y=np.array([10,35,25,19])
mylabels=["Cherries","Apples","Bananas","Dates"]
myexplode=[0,0.95,0.8,1]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([10,35,25,19])
mylabels=["Cherries","Apples","Bananas","Dates"]
plt.pie(y)
plt.legend(y,labels=mylabels)
plt.show()
```
